## **3. Data Cleaning:** Handle missing data by suitable method with explain why you use it.
---

According to basic statistics analysis, there are negative values in some columns (BasePay, TotalPay, TotalPayBenefits ...), but salaries can't be negative, so this may be errors in dataset so it is better to remove any values we aren't sure about, because that could affect the quality of the results of the analysis.
there are also two columns with no valuse at all (Notes, Statuse), we have to remove them too.


Also we have to deal with missing values, we can fill it with the mean of its column or we can remove them, I choose to remove missing values in 'BasePay, OvertimePay, OtherPay' columns because the dataset is large and it will not be affected by removing them, but for 'Benefits' column there are 36163 missing value, I replaced them with the mean of that column.

```python
# Impute missing values in 'Benefits' column with mean
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
```

```python
# displaying negative values in BasePay column
df_negative_basePay = df[df.BasePay < 0]
print('The records with negative base pay values: ', df_negative_basePay.shape[0])
df_negative_basePay
```

```python
# Removing rows with negative values and missing values
df_nonegatives = df[df.BasePay >= 0]
df_nonegatives = df_nonegatives[df_nonegatives.OvertimePay >= 0]
df_nonegatives = df_nonegatives[df_nonegatives.OtherPay >= 0]
df_nonegatives = df_nonegatives[df_nonegatives.Benefits >= 0]
df_nonegatives = df_nonegatives[df_nonegatives.TotalPay >= 0]
df_nonegatives = df_nonegatives[df_nonegatives.TotalPayBenefits >= 0]
df_nonegatives.describe()
```

```python
# Check for catigorical data
agencies = df.Agency.value_counts()
jobTitles = df.JobTitle.value_counts()
print('The agencies in dataset :\n', agencies)
print('The JobTitles in dataset :\n', jobTitles)
```

```python
# Removing columns with missing values (Notes, Status) and columns with the same value for all records (Agency)
df_cleaned = df_nonegatives.drop(labels=['Notes', 'Status', 'Agency'], axis=1)
```

```python
# Removing rows with any other missing values
df_cleaned = df_cleaned.dropna()
df_cleaned.info()
```

## **4. Basic Data Visualization:** Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set up the figure and axes for subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 5))
fig.tight_layout(pad=5.0)

# Plot histogram for salary distribution
sns.histplot(df_cleaned['BasePay'], bins=30, kde=True, color='gray', ax=axes[0,0])
axes[0,0].set_title('BasePay Distribution')
axes[0,0].set_xlabel('BasePay')
axes[0,0].set_ylabel('Frequency')
sns.histplot(df_cleaned['OvertimePay'], bins=30, kde=True, color='blue', ax=axes[0,1])
axes[0,1].set_title('OvertimePay Distribution')
axes[0,1].set_xlabel('OvertimePay')
axes[0,1].set_ylabel('Frequency')
sns.histplot(df_cleaned['OtherPay'], bins=30, kde=True, color='orange', ax=axes[0,2])
axes[0,2].set_title('OtherPay Distribution')
axes[0,2].set_xlabel('OtherPay')
axes[0,2].set_ylabel('Frequency')
sns.histplot(df_cleaned['Benefits'], bins=30, kde=True, color='lightblue', ax=axes[1,0])
axes[1,0].set_title('Benefits Distribution')
axes[1,0].set_xlabel('Benefits')
axes[1,0].set_ylabel('Frequency')
sns.histplot(df_cleaned['TotalPay'], bins=30, kde=True, color='red', ax=axes[1,1])
axes[1,1].set_title('TotalPay Distribution')
axes[1,1].set_xlabel('TotalPay')
axes[1,1].set_ylabel('Frequency')
sns.histplot(df_cleaned['TotalPayBenefits'], bins=30, kde=True, color='green', ax=axes[1,2])
axes[1,2].set_title('TotalPayBenefits Distribution')
axes[1,2].set_xlabel('TotalPayBenefits')
axes[1,2].set_ylabel('Frequency')
```

```python
# Plot pie chart for department proportions
fig, axes = plt.subplots(figsize=(10, 7))
years_counts = df_cleaned['Year'].value_counts()
axes.pie(years_counts, labels=years_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
axes.set_title('Employee Distribution by Years')
plt.show()
```

```python
# making pie chart for employee destribution based on JobTitle
# but there are 2159 category in this column, we can't show it all in pie chart so I mearged the small categories together
jobtitle_categories = df_cleaned['JobTitle'].value_counts()
large_categories = jobtitle_categories[jobtitle_categories>1500]
len(jobtitle_categories), large_categories
```

```python
def mearge_smalle_categories(x):
  if x not in large_categories:
    x ='Others'
  return x

df_cleaned['JobTitleCategories'] = df_cleaned['JobTitle'].map(mearge_smalle_categories)
```

```python
df_cleaned.info()
```

```python
department_counts = df_cleaned['JobTitleCategories'].value_counts()
fig, ax = plt.subplots(figsize=(10, 7))
ax.pie(department_counts, autopct='%2.1f%%', startangle=90, colors=sns.color_palette('pastel'))
ax.set_title('Employee Distribution by JobTitle')
ax.legend(department_counts.index,
          title="Categories",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```
